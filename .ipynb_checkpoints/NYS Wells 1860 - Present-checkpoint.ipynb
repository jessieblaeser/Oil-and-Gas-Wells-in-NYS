## 1. Filtering to keep only active oil and gas wells
We have omitted all "confidential" and "dry hole" wells as well. Dry hole wells are those found to be "incapable of producing either oil or gas in sufficient quantities to justify completion as an oil or gas well." 

For more info, [click here](https://www.eia.gov/tools/glossary/index.php?id=dry) for the oil/gas well glossary.

**Steps:**
1. Filter out all well types except for oil and gas
2. Filter to only include active wells
3. Download CSV of data 
4. Output = active_all_32522.csv

```python
# Filtering out all well types EXCEPT for oil and gas wells 
df = df_raw
df = df[~df["mapsymbol"].isin(["CON"])]
df = df[~df["mapsymbol"].isin(["DH"])]
df = df[~df["mapsymbol"].isin(["SM"])]
df = df[~df["mapsymbol"].isin(["SMP"])]
df = df[~df["mapsymbol"].isin(["O"])]
df = df[~df["mapsymbol"].isin(["OP"])]
len(df)
```

```python
# Filtering to only active wells
active = df[df["wellstatus"].isin(["AC"])]
# Counting number of active wells
len(active)
```

## 3. Filtering to keep only plugged and abandoned wells in each county
* [Abandoned Wells](https://www.dec.ny.gov/energy/111211.html#:~:text=To%20address%20these%20threats%2C%20the,of%20aquifers%20and%20surface%20waters.): "Abandoned wells are unplugged wells (primarily oil or gas wells) that have not been operated and maintained in accordance with prevailing statute and regulation."
* Plugged wells: Wells that have been properly filled after use has completed. For more info from the DEC on its plugging initiative, [click here](https://www.dec.ny.gov/energy/111252.html)

We are filtering for abandoned and *plugged* wells because when combined, these two numbers represent all **known** abandoned wells in the state, including those that have been plugged, and those yet to be plugged. 

For more info, [cllick here](https://www.dec.ny.gov/energy/111211.html#:~:text=To%20address%20these%20threats%2C%20the,of%20aquifers%20and%20surface%20waters.) for the NY DEC's explanation of unplugged, abandoned wells in the state.

## 4. Assessing time between well completion and date plugged

We have learned through reporing that the recommended lifespan of a well is ~30 years. Here we are adding a new column that measures the lifespan of wells in New York, measuring time between a well's completion date and plug date (if it has been plugged at all).

**Steps:**
1. Convert date columns in df to datetime format
2. Create new column by subtracting the date the well was completed from when it was plugged. This gives you the number of days the well has been open. This is not a measurement of how long the well was active. 
3. Filter to remove all NaN values from new 'well_time' column (these values occur every time there is an NA value in either 'date_well_completed' or 'date_well_plugged') 
4. Convert 'well_time' column to string, remove 'days', convert back into number fomat 
5. Filter to only include positive time values (see Note below)
6. Create 'well_time_yr' variable to reflect days in well lifecylces to years 
7. Download data 
8. Output = timespan_32522.csv

**Conlusions**
* Median time to plug a well from the day it is drilled is 8,744.5 days, or roughly 24 years
* Mean time time to plug a well is 11,034 days, or roughly 30 years (right on target) 
* The longest lifecycle of a well in NYS that has already been plugged is 47,326 days, or nearly 130 years
    * We see that this is the 'Root Charles Beldock 0' well, which was completed in 1890

**Note:**
A spokesperson for the DEC explained the 52 negative "time" values in the dataset as follows: "Possible explanations for this include but are not limited to when a well is re-entered for further development sometime after the well is plugged or, in most cases, the wells were dry (no oil or gas discovered) and were not “completed” for production but were marked as completed after the plugging and restoration of the dry well was finished."

```python
# Converting dates to Pandas date format
df['date_well_completed'] = pd.DatetimeIndex(df['datewellcompleted'])
df['date_well_plugged'] = pd.DatetimeIndex(df['datewellplugged'])

# Double checking datatypes are datetimes: 
df['date_well_completed']
df['date_well_plugged']
```

```python
# Make new variable subtracting the time between plugged and completed
df['well_time'] = df['date_well_plugged'] - df['date_well_completed']
```

```python
# Filtering to remove NAs that result from NAs in either 'date_well_plugged' or 'date_well_completed' variable
df_nona = df[~df['well_time'].isna()]
```

```python
df_nona['well_time']
```

```python
# Converting 'well_time' value (i.e, 5 days) to string to remove the word "days," so we are left with just the number '5' to convert to numeric form after
df_nona['well_time'] = df_nona['well_time'].astype(str)
df_nona['well_time'] = df_nona['well_time'].str.replace(' days', '')
```

```python
# Convert to float
df_nona['well_time'] = df_nona['well_time'].astype(float)
```

```python
# Filtering to only include positive times; 52 wells were omitted because they were recorded as plugged before completed.
df_pos_time = df_nona[df_nona['well_time']>0]
```

```python
# Looking at summary stats for the time in days between well completion and well plugging
df_pos_time['well_time'].describe()
```

```python
# Median time to plug a well from start of lifecylce: 
df_pos_time['well_time'].median()
```

```python
# Looking at distribution of 'well_time,' or distribution of days from the start of a well's life to when it was plugged
# For reference, 30 years translates to roughly 11,000 days

df_pos_time['well_time'].hist()
```

```python
# Taking a closer look at the outlier, maximum value
df_pos_time[df_pos_time['well_time']==47326]
```

```python
# Creating a new variable with WellTime as years instead of days
df_pos_time['well_time_yr'] = df_pos_time['well_time']/365
```

```python
# Looking at summary stats for the time in years between well completion and well plugging
df_pos_time.well_time_yr.describe()
```

## 5. Find means and other summary statistics for years between well completion and plugging by county

Here, we are creating a csv that breaks down everything we did above (finding summary stats for each plugged wells' lifespan) by county. This file will allow us to assess plugging time on a county level. 

**Steps:**
1. Group df by county
2. Grab summary stats for 'well_time_yr' for each county
3. Download data
4. Output = 'timespan_summ_32522.csv'


```python
#Find means and other summary statistics for years between well completion and plugging in each county
df_cty_summ = df_pos_time.groupby(['county']).well_time_yr.describe().reset_index()
df_cty_summ.sort_values('mean', ascending=False)
```
